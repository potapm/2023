```python
import os

input_file = "./data3/KRA-14-45-2023-08-31-morning.mp4"
output_file = "./temp.mp4"

os.system(f"ffmpeg -i {input_file} -s 512x288 {output_file}")
```

```python
%%time
import cv2
import numpy as np
cap = cv2.VideoCapture("temp.mp4")

ret, frame1 = cap.read()
prvs = cv2.cvtColor(frame1,cv2.COLOR_BGR2GRAY)
hsv = np.zeros_like(frame1)
hsv[...,1] = 255



# get video dims
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
size = (frame_width, frame_height)

# Define the codec and create VideoWriter object
#fourcc = cv2.VideoWriter_fourcc(*'DIVX')
fourcc = cv2.VideoWriter_fourcc(*'MP4V')
out = cv2.VideoWriter('./videoOFV2.mp4', fourcc, 20.0, size)

count=1

mspeed_M=[]

while cap.isOpened():
    ret, frame2 = cap.read()
    count += 5 # i.e. at 30 fps, this advances one second
    cap.set(cv2.CAP_PROP_POS_FRAMES, count)
    
    next = cv2.cvtColor(frame2,cv2.COLOR_BGR2GRAY)
    
    flow = cv2.calcOpticalFlowFarneback(prvs,next, None, 0.5, 3, 15, 3, 5, 1.2, 0)
    mspeed=np.absolute(flow).mean()
     

    
    mspeed_M.append(mspeed)
    if len(mspeed_M)>30:
        mspeed_M.pop(0)

    speed_rez=20*sum(mspeed_M) / len(mspeed_M) 
    
    mag, ang = cv2.cartToPolar(flow[...,0], flow[...,1])
    hsv[...,0] = ang*180/np.pi/2
    hsv[...,2] = cv2.normalize(mag,None,0,255,cv2.NORM_MINMAX)
    rgb = cv2.cvtColor(hsv,cv2.COLOR_HSV2BGR)

    
    font                   = cv2.FONT_HERSHEY_SIMPLEX
    bottomLeftCornerOfText = (30,30)
    fontScale              = 0.4
    fontColor              = (255,255,255)
    thickness              = 1
    lineType               = 2
    cv2.putText(rgb,str(speed_rez),
        bottomLeftCornerOfText,
        font,
        fontScale,
        fontColor,
        thickness,
        lineType)
    
    
    out.write(rgb)
    
    prvs = next
    print('.')


        
```

```python
cap.release()
out.release()
```
